```python
from PIL import Image
import cv2
from google.colab.patches import cv2_imshow
classifier=cv2.CascadeClassifier("haarcascade_frontalface_default.xml")

img =cv2.imread('/content/IIT-Bhilai-Day 2/Training_Friends_Dataset_100_imgs/04f48620-de82-4f55-a787-ac234cd861d2.JPG')

gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

face=classifier.detectMultiScale(gray)

for (x,y,w,h) in face:
  cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,225),3)
  roi_color = img[y:y + h, x:x + w]
  print("[INFO] Object found. Saving locally.")
  cv2.imwrite(str(w) + str(h) + '_faces.jpg', roi_color)

cv2_imshow(img)
```

```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
img = cv.imread('/content/111111_faces.jpg',0)
img2 = img.copy()
template = cv.imread('/content/IIT-Bhilai-Day 2/Training_Friends_Dataset_100_imgs/3b46cbe1-7a6e-4dd7-910e-4705407f7724.JPG',0)
w, h = template.shape[::-1]
# All the 6 methods for comparison in a list
methods = ['cv.TM_CCOEFF', 'cv.TM_CCOEFF_NORMED', 'cv.TM_CCORR',
            'cv.TM_CCORR_NORMED', 'cv.TM_SQDIFF', 'cv.TM_SQDIFF_NORMED']
for meth in methods:
    img = img2.copy()
    method = eval(meth)
    # Apply template Matching
    res = cv.matchTemplate(img,template,method)
    min_val, max_val, min_loc, max_loc = cv.minMaxLoc(res)
    # If the method is TM_SQDIFF or TM_SQDIFF_NORMED, take minimum
    if method in [cv.TM_SQDIFF, cv.TM_SQDIFF_NORMED]:
        top_left = min_loc
    else:
        top_left = max_loc
    bottom_right = (top_left[0] + w, top_left[1] + h)
    cv.rectangle(img,top_left, bottom_right, 255, 2)
    plt.subplot(121),plt.imshow(res,cmap = 'gray')
    plt.title('Matching Result'), plt.xticks([]), plt.yticks([])
    plt.subplot(122),plt.imshow(img,cmap = 'gray')
    plt.title('Detected Point'), plt.xticks([]), plt.yticks([])
    plt.suptitle(meth)
    plt.show()
```

```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
img_rgb = cv.imread('/content/IIT-Bhilai-Day 2/Training_Friends_Dataset_100_imgs/3b46cbe1-7a6e-4dd7-910e-4705407f7724.JPG')
img_gray = cv.cvtColor(img_rgb, cv.COLOR_BGR2GRAY)
template = cv.imread('/content/111111_faces.jpg',0)
w, h = template.shape[::-1]
res = cv.matchTemplate(img_gray,template,cv.TM_CCOEFF_NORMED)
threshold = 0.8
loc = np.where( res >= threshold)
for pt in zip(*loc[::-1]):
    cv.rectangle(img_rgb, pt, (pt[0] + w, pt[1] + h), (0,0,255), 2)
cv.imwrite('res.png',img_rgb)
cv2_imshow(img_rgb)
```

```python
import os


imlable=[]
for fname in os.listdir('/content/IIT-Bhilai-Day 2/Training_Friends_Dataset_100_imgs/'):
  imlable.append(fname)
  imlable.sort()
print(imlable)
```

```python
sol1=[]
sol2=[]
sol=[]

for face in imlable:
   img_rgb = cv2.imread('/content/5757_faces.jpg')
   img_gray = cv2.cvtColor(img_rgb, cv2.COLOR_BGR2GRAY)
   template = cv2.imread('/content/IIT-Bhilai-Day 2/Training_Friends_Dataset_100_imgs/'+face, 0)
   w, h = template.shape[::-1]
   res = cv2.matchTemplate(img_gray, template, cv2.TM_CCOEFF_NORMED)
   threshold = 0.8
   loc = np.where(res >= threshold)
   if res.all()>=threshold:
     sol.append("yes")
     sol1.append("yes")
   else:
     sol.append("no")
     sol2.append("no")

   for pt in zip(*loc[::-1]):
     cv2.rectangle(img_rgb, pt, (pt[0] + w, pt[1] + h), (0, 255, 255), 2)

print(sol)
print(sol1)
print(sol2)
```
