```python
# extract and plot each detected face in a photograph
import os
import cv2
from matplotlib import pyplot
from matplotlib.patches import Rectangle
from matplotlib.patches import Circle
from mtcnn.mtcnn import MTCNN

PATH = "/content/IIT-Bhilai-Day 2/Training_Friends_Dataset_100_imgs/"
k=0
images = []
image_labels = []
for filename in os.listdir(PATH):
  # load image from file
  pixels = pyplot.imread(PATH+filename)
  # create the detector, using default weights
  detector = MTCNN()
  # detect faces in the image

  faces = detector.detect_faces(pixels)
  for i in range(len(faces)):
    data = cv2.imread(PATH+filename)
    x1, y1, width, height = faces[i]['box']
    x2, y2 = x1 + width, y1 + height
    image = data[y1:y2, x1:x2]
    images.append(image)
    image_labels.append(str(i)+'_'+str(k))
k=k+1
```

```python
import os

names=[]
for fname in os.listdir("/content/IIT-Bhilai-Day 2/Training_Friends_Dataset_100_imgs/"):
  names.append(fname)
names.sort()
print(names)
print(len(names))
```

```python
from PIL import Image
import cv2
from google.colab.patches import cv2_imshow
classifier=cv2.CascadeClassifier("haarcascade_frontalface_default.xml")

for i in names:
  img =cv2.imread('/content/IIT-Bhilai-Day 2/Training_Friends_Dataset_100_imgs/'+i)

  gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

  face=classifier.detectMultiScale(gray)

  for (x,y,w,h) in face:
    cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,225),3)
    roi_color = img[y:y + h, x:x + w]
    print("[INFO] Object found. Saving locally.")
    cv2.imwrite('/content/faces/'+str(w) + str(h) + '_faces.jpg', roi_color)


```

```python
flable=[]
for fname in os.listdir('/content/faces/'):
  flable.append(fname)
flable.sort()
print(flable)
```

```python
sol=[]
sol1=[]

for face in flable:
   img_rgb = cv2.imread('/content/faces/103103_faces.jpg')
   img_gray = cv2.cvtColor(img_rgb, cv2.COLOR_BGR2GRAY)
   template = cv2.imread('/content/faces/'+face, 0)
   w, h = template.shape[::-1]
   res = cv2.matchTemplate(img_gray, template, cv2.TM_SQDIFF)
   threshold = 0.8
   loc = np.where(res >= threshold)
   if res.all()>=threshold:
     sol.append("yes")
   else:
     sol.append("no")
     sol1.append("no")

   for pt in zip(*loc[::-1]):
     cv2.rectangle(img_rgb, pt, (pt[0] + w, pt[1] + h), (0, 255, 255), 2)

print(sol)
print(sol1)
```
